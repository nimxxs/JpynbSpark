### 간단한 시각화
* plot(x축, y축, 옵션)
* 컬럼명 : 객체명['컬럼명'], 객체명.컬럼명

```python
# FutureWarning, UserWarning 감추기
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
warnings.simplefilter(action='ignore', category=UserWarning)
```

```python
# 옵션을 안달면 그냥 선으로 나옴
plt.plot(iris2.sepal_length, iris2.sepal_width)
```

```python
plt.plot(iris2.sepal_length, iris2.sepal_width, 'o')
```

```python
plt.plot(iris2.sepal_length, iris2.sepal_width, 'ro')
plt.tight_layout()  # 젤 위에 나오는 문자들 없애줌
```

```python
# scatterplot : 산점도
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris2)
plt.tight_layout()
```

```python
sns.pairplot(iris2, x_vars='sepal_length', y_vars='sepal_width', hue='species', height=5)
plt.tight_layout()
```

```python
#  상관계수
iris.corr('sepal_length','sepal_width')
```

```python
iris.corr('petal_length','petal_width')
```

```python
iris2.iloc[:, 0:4].corr()
```

### tip 데이터 EDA

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# summary
tips.describe()
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='day', data=tips)
plt.tight_layout()
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='time', data=tips)
plt.tight_layout()
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='size', data=tips)
plt.tight_layout()
```

```python
# 빈도를 표시해주는 그래프
sns.countplot(x='sex', data=tips)
plt.tight_layout()
```

```python
# 빈도를 표시해주는 그래프
sns.countplot(x='day', data=tips)
plt.tight_layout()
```

```python
# 빈도를 표시해주는 그래프
sns.countplot(x='size', data=tips)
plt.tight_layout()
```
